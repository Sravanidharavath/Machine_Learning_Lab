```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
print(df['Species'].unique())  # Now you should see [0 1 2] if 3 classes exist
```

```python
#19.prepare x and y
x=df.drop(columns=['Species','Id'])
y=df['Species']
print(x.shape)
print(y.shape)
```

```python
#20.split the traina dn test data 70 %train 30%test random state=1
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#21.linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.intercept_)
print(model.coef_)
```

```python
#22.decision tree
from sklearn.tree import DecisionTreeClassifier,plot_tree
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
print(dt.score(xtest,ytest))
plot_tree(dt)
plt.show()
```

```python
#24.knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
knn.score(xtest,ytest)*100
```

```python

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)


# Initialize the model
model = SVC(kernel='linear', C=1.0)

# Fit the model
model.fit(xtrain, ytrain)

# Predict using the model
predictions = model.predict(xtest)
print(predictions)

# Calculate accuracy
accuracy = accuracy_score(ytest, model.predict(xtest))
print(f"Accuracy of the model: {accuracy * 100:.2f}%")
```

```python
#26.Evalutae the model using  confusion matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix,f1_score,recall_score,classification_report,accuracy_score,precision_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
print(accuracy_score(ytest,dt.predict(xtest)))
print(f1_score(ytest,dt.predict(xtest),average="weighted"))
print(recall_score(ytest,dt.predict(xtest),average='weighted'))
print(precision_score(ytest,dt.predict(xtest),average='weighted'))
print("\nClassification Report:\n", classification_report(ytest, dt.predict(xtest)))
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
print(dt.score(xtest,ytest))
plot_tree(dt)
plt.show()
cf=confusion_matrix(ytest,dt.predict(xtest))
sns.heatmap(cf,annot=True)
plt.show()
```

```python
#k-fold cross validation
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Create the model
model = SVC(C=1, kernel='linear')

# 5-fold cross-validation
scores = cross_val_score(model, x, y, cv=5)

# Results
print("Cross-validation scores for each fold:", scores)
print("Average cross-validation score:", scores.mean())
```
